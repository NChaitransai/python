```python
lst=list(map(int,input().split()))
even=[]
odd=[]
for i in lst:
    if i%2==0:
        even.append(i)
else:
    odd.append(i)
print('even list is', even)
print('odd list is', odd)
```

```python
#square pattern# * * * * *
# * * * * *
# * * * * *
# * * * * *
# * * * * *
for i in range(5):
    for j in range(5):
        print("*",end=" ")
    print()
```

```python
#upper right angled pattern
n=5
for i in range(n):
    for j in range(n):
        if i==0 or i==n-1 or j==0 or j==n-1:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
arr=list(map(int, input().split()))
target=int(input())
i=0
while i<len(arr):
    if arr[i]==target:
        del arr[i]
    else:
        i+=1
print(arr)
```
