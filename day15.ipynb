```python
n=7 
for i in range(n):
    for j in range(n): 
       if j==0 or j==n-1 or i==j : 
           print("*",end=" ")
       else:
           print(" ",end=" ")
    print()
```

```python

n=7 
for i in range(n):
    for j in range(n): 
       if (i+j)%2==0: 
           print("0",end=" ")
       else:
           print("1",end=" ")
    print()


```
