```python
#set operations
#union
#intersection
s1={1,2,3,6}
s2={3,4,5,6}
print(s1.union(s2))
print(s1 |s2)
print("**************intersection***********")
print(s1.intersection(s2))
print(s1&s2)
print("***************difference***************")
print(s1.difference(s2))
print(s1-s2)
print("***************symmetric difference***************")
print((s1).symmetric_difference(s2))                     
print("********subset************")
print({1,2}.issubset(s1))
print({1,2,5}.issubset(s1))
print("***********superset***********")
print(s2.issuperset(s1))
print(s1.issuperset({1,2}))
print({1,2,5}.issuperset(s1))
print("*****disjointset***************")
print(s1.isdisjoint(s2))
print(s1.isdisjoint({10}))

```

```python
#add(ele): It add to set, if the element is not present in set , otherwise nothing 
#update(new_set): set element updates with new set elements
#remove(ele): It removes the set element, if element not present in set it raises key
#discard(ele): It removes the ele from set otherwise nothing. 
s = {1,2,3,4}
s.add(10)
print(s)
s.update({1,2,5,6})
print(s)
# s.remove(10)   key error
# print(s)
s.discard(10)
print(s)
# s.remove(20)  key error
# print(s)
s.discard(20)
print(s)
```
