```python
password = input()

upper = lower = digit = special = False

for ch in password:
    if ch.isupper():
        upper = True
    elif ch.islower():
        lower = True
    elif ch.isdigit():
        digit = True
    elif ch in  '@#$^!':   
        special = True

if len(password) >= 8 and upper and lower and digit and special:
    print("Strong password")
else:
    print("Weak password")
```
