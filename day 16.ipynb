```python
#dictionary is ordered collection
details={'name' :"chaitran","Batch":49,"course": "python"}
print(details)
         
```

```python
#dictionary key are unique
details={'name' :"chaitran","Batch":49,"course": "python",'name': "sai"}
print(details)
         
```

```python
#dictionary is ordered collection
details={'name' :"chaitran","Batch":49,"course": "python",'name': "sai"}
print(details['name'])
         
```

```python
# dictionary doesnot support indexing or slicing
details={'name' :"chaitran","Batch":49,"course": "python",'name': "sai"}
print(details[0])
print(details[1])
```

```python
#dictionary is mutable
details={'name' :"chaitran","Batch":49,"course": "python",'name': "sai",0:1}
details['course']='java'
details['age']= 50
print(details)
```

```python
#dictionary methods
#get(key,default_values) 
details={'name' :"chaitran","Batch":49,"course": "python",'name': "sai",0:1}
print(details.get('location','india')) 
print(details.get('name','ram'))
print(details.get('marks'))
```

```python
#update(items)
details={'name' :"chaitran","Batch":49,"course": "python",'name': "sai",0:1}
new_dict={'course':'java','Batch': 50, 'location' :'india'}
details.update(new_dict)
print(details)
```

```python
#keys()
#values()
#items()
details={'name' :"chaitran","Batch":49,"course": "python",'name': "sai",0:1}
print(details.keys())
print(details.values())
print(details.items())
print(li+st(details.items()))
```
