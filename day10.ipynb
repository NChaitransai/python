```python
def lst(x):
    sum=0
    for i in x:
        sum+=i
    return sum
res=lst([1,2,3,4,5])
print(res)
```

```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mult(a,b):
    return a*b
def div(a,b):
    if b==0:
        return 'cannot divide by zero'
    else:
        return a/b

print(add(49,73))
print(sub(12,13))
print(mult(12,13))
print(div(12,0))
```
